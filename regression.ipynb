```python

import seaborn as sns
def visualizeRelations(data):
    print()
    print('Relationship between age, Sex and Haemoglobin_level: ')
    sns.barplot(x="Age", y="Haemoglobin_level", hue="Sex", data=data)
    plt.show()
   
  
    print('Analyzing the relationship between gender and Haemoglobin_level')
    pie = plt.figure(figsize=(25,10))
    pie.add_subplot(121)
    data.Haemoglobin_level[data['Sex'] == 'Male'].value_counts().plot(kind='bar')
    pie.add_subplot(122)
    data.Haemoglobin_level[data['Sex'] == 'Female'].value_counts().plot(kind='bar')
    plt.show()
    
    
    
    print()
    print()
    print('Visualize correlation between some features: ')
    print()
    correlation_matrix = data[data.columns.to_series().sample(20)].corr()
    plt.figure(figsize=(10,8))
    ax = sns.heatmap(correlation_matrix, vmax=1, square=True, annot=True,fmt='.2f', cmap ='GnBu', cbar_kws={"shrink": .5}, robust=True)
    plt.title('Correlation matrix between the features', fontsize=20)
    plt.show()
    print()
    
visualizeRelations(dt)
```
