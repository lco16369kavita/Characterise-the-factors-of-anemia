```python
# To change Age months into years so we create a dataset

# Copy dataframe df1 to df2 where age _code in months
df2 = df1[df1['Age_Code'] =='Months']
```

```python
#Check df2
df2[['Age','Age_Code']]
```

```python
#Convert Age months to years
df2['Age']=df2['Age']/12
```

```python
#Create a new dataframe df3 where age_code is year And then concatenate with df2
df3=df1[df1['Age_Code'] =='Year']
```

```python
#Concatenate df2,df3 and make a new dataframe df4 and in this the age in years
frames=[df2, df3]
df4 = pd.concat(frames)
```
